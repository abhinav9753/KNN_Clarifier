```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

df = pd.read_csv('/content/Iris.csv')
df
```

```python
df = df.drop(['Id'], axis = 1)
df
```

```python
df['Species'].value_counts()
```

```python
df['Species'] = df['Species'].map({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
df
```

```python
X = df.iloc[: , :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.5, random_state=42)
```

```python
len(X_train), len(X_test)
```

```python
len(y_train), len(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
reg = KNeighborsClassifier(n_neighbors=5)
reg.fit(X_train, y_train)
```

```python
y_train_pre = reg.predict(X_train)
traing_data = pd.DataFrame()
traing_data = X_train
traing_data['y_train'] = y_train
traing_data['y_train_prediction'] = y_train_pre
traing_data
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_train, y_train_pre)
```

```python
accuracy_score(y_train, y_train_pre)
```

```python
print(classification_report(y_train, y_train_pre))
```

```python
y_test_prediction = reg.predict(X_test)
test_data = pd.DataFrame()
test_data = X_test
test_data['y_test'] = y_test
test_data['y_test_prediction'] = y_test_prediction
test_data
```

```python
confusion_matrix(y_test, y_test_prediction)
```

```python
accuracy_score(y_test, y_test_prediction)
```

```python
#testing own data
import warnings
warnings.filterwarnings('ignore')
big_len = 3.4
big_wid = 4.7
small_len = 1.9
small_wid = 0.2

if reg.predict([[big_len, big_wid, small_len, small_wid]]) == 0:
  print("Setosa")
elif reg.predict([[big_len, big_wid, small_len, small_wid]]) == 1:
  print("Virsica")
else:
  print("Virginica")
```

##Cross validation

```python
from sklearn.model_selection import cross_val_score
knn_reg = KNeighborsClassifier(n_neighbors=5)

test_acc = cross_val_score(knn_reg, X, y, cv = 5)
print(f'Testing avg accuracy values :{np.mean(test_acc)}')
```
